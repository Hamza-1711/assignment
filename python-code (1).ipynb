```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df= "/kaggle/input/natural-disaster-data"
```

```python


df = pd.DataFrame(df)

# Display histogram
def plot_histogram(dataframe, column):
    plt.hist(dataframe[column], bins=10, color='blue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

plot_histogram(df, 'Magnitude')

# Display pie chart
def plot_pie_chart(dataframe, column):
    plt.figure(figsize=(8, 8))
    plt.pie(dataframe[column].value_counts(), labels=dataframe[column].unique(), autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart of {column}')
    plt.show()

plot_pie_chart(df, 'Year')


# Display correlation heatmap
def plot_heatmap(dataframe):
    correlation_matrix = dataframe.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap')
    plt.show()

plot_heatmap(df)





# Display describe statistics
describe_stats = df.describe()
print(describe_stats)
```
